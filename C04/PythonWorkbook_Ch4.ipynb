```python
import token
from operator import countOf

astring = 'abcdefghijklmnopqrstuvwxyz'
alist = [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]
atuple = (0, 2, 4, 6, 8, 10, "x", "y", "z")
adict = {}


def first_last(seq: str | list | tuple) -> str | list | tuple:
    newlst = []
    if isinstance(seq, str):
        newstr = f'{seq[0]}{seq[-1]}'
        return newstr
    elif isinstance(seq, list):
        newlst.append(seq[0])
        newlst.append(seq[-1])
        return newlst
    elif isinstance(seq, tuple):
        newtuple = seq[0], seq[-1]
        return newtuple
    else:
        return 'None'


print(first_last(astring))
print(first_last(alist))
print(first_last(atuple))
```

## Exercise 10: Summing anything

```python
def my_sum(*args):
    if not args:
        return args
    output = args[0]
    for _ in args[1:]:
        output += _
    return output


print(my_sum())
print(my_sum(10, 20, 30, 40))
print(my_sum('a', 'b', 'c', 'd'))
print(my_sum([10, 20, 30], [40, 50, 60], [70, 80]))
print(my_sum((10, 20, 30), (40, 50, 60), (70, 80)))
# print(my_sum({10, 20, 30}, {40, 50, 60}, {70, 80}))
```

```python
def mysum_bigger_than(*items):
    """Sum items, which should be of the same type.
Ignore any below the value of threshold.
The arguments should handle the + operator.
If passed no arguments, then return an empty tuple.
"""


def mysum_bigger_than(*items):
    if len(items) < 2:
        return items
    threshold = items[0]
    first_item = items[1] if len(items) > 1 else threshold
    output = first_item[:0] if hasattr(first_item, '__getitem__') else 0
    for item in items[1:]:
        if item > threshold:
            output += item
    return output


print(mysum_bigger_than(10, 5, 20, 30, 6))
print(mysum_bigger_than("b", "a", "x", "y", "z"))
```

```python
def mysum_bigger_than(threshold, *items):
    """Sum items, which should be of the same type.
Ignore any below the value of threshold.
The arguments should handle the + operator.
If passed no arguments, then return an empty tuple.
"""
    if not items:
        return items
    output = 0
    for item in items:
        if item > threshold:
            output += item
    return output


print(mysum_bigger_than(10, 5, 20, 30, 6))
print(mysum_bigger_than("b", "a", "x", "y", "z"))
```

```python
def sum_numeric(*items):
    """Sum all items, assuming that they
are integers or can be turned into integers.
"""
    total = 0

    for item in items:
        try:
            total += int(item)
        except ValueError:
            pass
    return total


print(sum_numeric(10, 20, 'a', '30', 'abc'))
```

```python
def dictlst(*args: dict) -> dict:
    superd = {}
    for arg in args:
        for k, v in arg.items():
            if k in superd:
                try:
                    superd[k].append(v)  # if value is a list append it
                except AttributeError:
                    superd[k] = [superd[k], v]  # if value is not a list make it a list and add both values
            else:
                superd[k] = v
    return superd


dicta = {"a": "a"}
dictb = {"a": "b", "b": "a"}
dictc = {"a": "c", "b": "b", "c": "a"}
print(dictlst(dicta, dictb, dictc))
```

```python
# Lerner's
from collections import Counter

WORDS = ['this', 'is', 'an', 'elementary', 'test', 'example']


def most_repeating_letter_count(word):
    """Given a non-empty string, counts how
many times each letter appears in the string,
and returns an integer indicating how often
the most common letter appears."""
    return Counter(word).most_common(1)[0][1]


def most_repeating_word(words):
    """Given a list of non-empty strings (words),
returns the word containing at least one letter that repeats
more often than any letter in any other word.

Because sorting in Python is stable, if multiple words have
the same count, then the first will be returned."""
    return max(words, key=most_repeating_letter_count)


print(most_repeating_word(WORDS))
```

```python
# Lerner
from collections import defaultdict


def shells_and_users_by_popularity(filename):
    shells = defaultdict(list)
    for one_line in open(filename):
        if one_line.startswith(('#', '\n')):
            continue

        username, *rest, shell = one_line.strip().split(':')

        shells[shell].append(username)

    return sorted(shells.items(), key=len)

print(shells_and_users_by_popularity("../linux-etc-passwd.txt"))
```

## Exercise 13: Printing tuple records

```python
import operator
COUNTRIES = [('Canada', 9984670, 38250000),
             ('Italy', 301340, 59110000),
             ('United Kingdom', 242495, 67220000),
             ('France', 551695, 67390000),
             ('Germany', 357022, 83200000),
             ('Japan', 377975, 125700000),
             ('United States', 9833517, 331900000)
            ]
def format_sort_records(lst: list(tuple)) -> list:
    output = []
    for item in sorted(lst, key=operator.itemgetter(0)):
        output.append(f'{item[0]:15} {item[1]:10,} {item[2]:15,}')
    return output
print('\n'.join(format_sort_records(COUNTRIES)))
```
